```python
import pandas as pd
from datetime import datetime
from datetime import date
import numpy as np
```

```python
df = pd.read_csv('Data/bank_transactions.csv',usecols=['CustomerID','CustomerDOB','CustGender','CustLocation','CustAccountBalance'])
```

```python
df=df.groupby('CustomerID').first().reset_index()
```

```python
df['CustomerDOB']= pd.to_datetime(df['CustomerDOB'])
```

```python
df.drop(df[(df['CustomerDOB'] == '1800-1-1')].index, inplace = True)
```

```python
df=df.dropna()
```

```python
def age(birthdate):
    today = date.today()
    one_or_zero = ((today.month, today.day) < (birthdate.month, birthdate.day))
    year_difference = today.year - birthdate.year
    age = year_difference - one_or_zero
    return age
```

```python
df["CustomerDOB"] = df['CustomerDOB'].apply(age)
```

```python
df.drop(df[(df['CustomerDOB'] < 16)].index, inplace = True)
```

```python
df['CustGender']=df['CustGender'].apply(lambda x: 1 if (x=='M') else 0)
```

```python
def ageClassification(age):
    if age <= 21:
        return 1
    elif age > 21 and age <= 30:
        return 2
    elif age > 30 and age <= 45:
        return 3
    elif age > 45 and age <= 60:
        return 4
    else:
        return 5    
```

```python
df['CustomerDOB']=df['CustomerDOB'].apply(ageClassification)
```

```python
location_array = df['CustLocation'].unique()
location_dictionary={}
counter = 0
for i in location_array:
    location_dictionary[i]=counter
    counter+=1
```

```python
df['CustLocation']=df['CustLocation'].apply(lambda x: location_dictionary[x])
```

```python
df
```

```python
a = pd.qcut(np.array(df["CustAccountBalance"]), 4)
```

```python
a = a.unique().dropna()
```

```python
def balanceClassification(x):
    if x in a[1]:
        return 1
    elif x in a[3]:
        return 2
    elif x in a[0]:
        return 3
    else:
        return 4
```

```python
df["CustAccountBalance"] = df['CustAccountBalance'].apply(balanceClassification)
```
